```python
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"
```

```python
df_Nikkei["Trend"]= df_Nikkei["Day_Perc_Change"].apply(lambda x:trend(x))
df_Nasdaq["Trend"]= df_Nasdaq["Day_Perc_Change"].apply(lambda x:trend(x))
df_HangSeng["Trend"]= df_HangSeng["Day_Perc_Change"].apply(lambda x:trend(x))
```

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
# type your code here
# load stock market data
df_Nikkei = pd.read_csv("Nikkei.csv") 
df_Nasdaq = pd.read_csv("Nasdaq.csv") 
df_HangSeng = pd.read_csv("Hang_Seng.csv") 

# merge Adj. close values
combined_df = pd.concat([df_Nikkei["Adj Close"], df_Nasdaq["Adj Close"], df_HangSeng["Adj Close"]], axis=1)
combined_df.columns = ["Nikkei", "Nasdaq", "Hang_Seng"]
combined_df.head()
```

```python
# drop null values
combined_df.dropna(inplace = True, axis = 0)

# display first few rows
combined_df.head()
```

```python
#membandingkan data sekarang dengan data sebelumnya kemudian dikalikan dengan 100. lalu disimpan ke dataframe yang baru
pct_chg_df = combined_df.pct_change()*100

#drop baris yang terdapat null value
pct_chg_df.dropna(inplace = True, how = "any", axis = 0)

#menampilkan beberapa baris data terkini
pct_chg_df.head()
```

```python
pct_chg_df.corr()
```

visualisai

```python
volatility = pct_chg_df[['Nikkei', 'Nasdaq', 'Hang_Seng']].rolling(7).std()*np.sqrt(7)
sns.set_theme(context='poster')
sns.relplot(data=volatility,kind="line",height=5, aspect=2.5,dashes = False).set(title="Volatilitas dari Ketiga Perusahaan")
```

#### Penjelasan 
Dari volatilitas diatas dapat dilihat bahwa pada saat 300-400 harga saham nasdaq sangat tinnggi dengan pemerosotan saham hangseng, sedangkan pada saat 500 semua harga saham mengalami penurunan drastis yang dimana harga saham sampai 0.
